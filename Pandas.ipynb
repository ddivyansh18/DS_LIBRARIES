```python
data = ('index,a,b,c\n'
       '4,apple,bat,5.7\n'
       '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data))
```

```python
#Make 0th column as index
pd.read_csv(StringIO(data),index_col=0)
```

```python
data = ('a,b,c\n'
       '4,apple,bat,\n'
       '8,orange,cow,')
pd.read_csv(StringIO(data))
```

```python
#In this case 4 and 8 were not index
pd.read_csv(StringIO(data),index_col=False)
```

```python
#Combine usecols and Index_col
data = ('a,b,c\n'
       '4,apple,bat,\n'
       '8,orange,cow,')
pd.read_csv(StringIO(data),usecols=['b','c'],index_col=False)
```
