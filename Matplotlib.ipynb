```python
import matplotlib.pyplot as plt
%matplotlib inline
#Then no need to write plt.show()
```

```python
import numpy as np
```

```python
x = np.arange(10)
y = np.arange(11,21)
```

```python
#Scatterplot is used to scatter values in a 2d plane
plt.scatter(x,y,c='g') # g is green
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Graph')
#We can save by right clicking or
plt.savefig('Test.png')
```

```python
#In plot we will be getting a straight line
y = x*x
plt.plot(x,y,'r*')
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'r',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Diagram')
```

```python
#Creating subplots
#Multiple plots in a single plot
#rows,col,first pos
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'mo-')
```

```python
x = np.arange(1,11)
y = 5 * x + 3
plt.title('plt graph')
plt.xlabel('x')
plt.ylabel('y=5x+3')
plt.plot(x,y)
```

```python
np.pi
```

```python
#Plotting a sine curve
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title('Sine wave')
plt.plot(x,y,'ro')
```

```python
#Bar plot
x = [2,8,10]
y = [11,16,9]
x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
#Histogram
a = np.array([2,2,2,3,4,5,6,6,7,7,8,9,3,4,5,7])
#In x axis we will have given values, and in y axis we will have dinsity 
#i.e count of that value
#By default there are 10 bins
plt.hist(a)
plt.title('Histogram')
```

```python
plt.hist(a,bins=20)
```

```python
#Box plot
#Helps to find percentiles and detect outliers
data = [np.random.normal(0,std,100) for std in range(1,4)]
data
```

```python
#vert is for making it vertically
#patch_artist enables the colour in boxes
plt.boxplot(data,vert=True,patch_artist=True)
```

```python
#Pie chart
labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0) #Explode first slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct ='%1.1f%%', shadow=True)
plt.axis('equal')
```
