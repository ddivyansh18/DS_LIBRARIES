- Whenever we get a dataset suppose if we have features
 like f1,f2,f3 .. then we have to divide them into
 dependent and independent features
- We can easily plot upto 3d diagrams, but for more than
 three features, it becomes very difficult to visulaise.
- If we consider only one variable for analysis
 out of all then it is called univariate analysis. 
- Analysis of two features is called bivariate analysis
- Joinplot is used for bivariate analysis
- Pairplot is used for analysis of more than two features


# Non-categorical Plots
- Heatmap
- Jointplot
- Pairplot

```python
import seaborn as sns
```

```python
#Using inbuilt tips dataset
df = sns.load_dataset("tips")
```

```python
df.head()
```

- Suppose we want to create a model to predict tips based on other features
- Here tip is a dependent feature
- Other features are independent features

# Correlation with Heatmap
- It shows a 2D correlation matrix between two discrete dimensions, important for
 feature selection. 
- Only works for int and float
- Value of pearson's correlation coefficient varies between -1 to 1
- Higher positive value means directly proportional
- Higher negative value means inversely proportional
- Zero means no correlation

```python
#Only three features as those are only int and float
df.corr()
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())
```

# Jointplot
- Only used with two features
- Used for univariate or bivariate analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

- Most people have a total_bill of 10 units and tip of 2 units

```python
#Regression - best fit line
#ALso plots pdf(probability density function)
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pairplot
- Also called as scatterplot
- One variable in same row is matched with another variable's value

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

# Distplot
- Used to create histogram
- Used to check the distribution of features

```python
sns.distplot(df['tip'])
```

```python
#KDE is Kernel Density Estimate - Used for non parametric analysis
#If KDE is True it shows the density
#If KDE is False it shows the count
sns.distplot(df['tip'],kde=False,bins=10)
```

# Categorical Plots
- boxplot
- violinplot
- countplot
- barplot

# Countplot

```python
sns.countplot("smoker",data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y="sex",data=df)
```

# Barplot
- We have to give both x and y axis values
- Whereas in countplot we gave only one asix

```python
sns.barplot(x="total_bill",y="smoker",data=df)
```

```python
sns.barplot(x="sex",y="total_bill",data=df)
```

# Boxplot
- Boxplot is also called as whiskerplot
- It is a graph that represents information from a five-number summary

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x="day",y="total_bill",data=df,palette="rainbow")
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# Violinplot
- Helps us to see data in terms of both Kernel Density Estimation and box plot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
